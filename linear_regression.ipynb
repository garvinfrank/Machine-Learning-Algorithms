```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
hp = pd.read_csv('/content/homeprices - homeprices.csv')
```

```python
hp.head(8)
```

```python
hp.describe()
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(hp.area,hp.price)
```

```python
x = hp.iloc[:,0:1]
display(x)
```

```python
y = hp.iloc[:,1:2]
display(y)
```

```python
model = linear_model.LinearRegression()
model.fit(x,y)
```

```python
model.predict([[2500]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
ar = pd.read_csv('/content/areas - areas.csv')
```

```python
ar['price'] = model.predict(ar)
display(ar)
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(hp.area,hp.price,color='green')
plt.scatter(ar.area,ar.price,color='red')
plt.plot(ar.area,ar.price)
```
