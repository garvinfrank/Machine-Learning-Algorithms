```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('Iris.csv')
data
```

```python
data['Species'].unique()
```

```python
data['Species']= data['Species'].replace({'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3})
```

```python
data
```

```python
X_train,X_test,y_train,y_test= train_test_split(data[['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']],data['Species'], test_size=0.3)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn import linear_model
model = linear_model.LogisticRegression(max_iter=200)
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
X_test
```

```python
model.score(X_test,y_test)
```

```python
predected_output = model.predict(X_test)
predected_output
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,predected_output)
cm
```

```python
import seaborn as sn
from matplotlib import pyplot as plt
plt.figure(figsize=(5,4))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')

plt.ylabel('Truth')
```

```python
from sklearn.svm import SVC
svm_model_linear = SVC(kernel = 'linear', C = 1).fit(X_train, y_train)
svm_predictions = svm_model_linear.predict(X_test)
```

```python
accuracy = svm_model_linear.score(X_test, y_test)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,svm_predictions)
cm
```
